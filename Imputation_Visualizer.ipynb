# **Imputation Visualizer using holy trio of statistics: Mean, Median, Mode**

```python
import matplotlib.pyplot as plt #LIBRARIES
import numpy as np
```

```python
plt.style.use("default")
```

This sets the plotting style to "default", which is the original matplotlib look.

Matplotlib has multiple built-in styles (like "ggplot", "seaborn", "bmh"), and you can switch between them to change the overall look of your plots easily.

Using "default" resets any custom styling back to the standard matplotlib appearance.

```python
plt.rcParams['axes.grid'] = True
```

plt.rcParams is a dictionary-like object that stores matplotlib configuration settings.

Setting axes.grid to True enables grid lines on all plots by default. So, every plot we create will show grid lines on the axes unless we override it.

```python
plt.rcParams['grid.alpha'] = 0.3
```

This sets the transparency (alpha) of the grid lines.

alpha ranges from 0 (fully transparent) to 1 (fully opaque).

So 0.3 makes the grid lines quite faint/light — you can still see them, but they don’t overpower the plot.

**Sample data for different Imputation Methods:**

```python
mean_data = [20, 22, 25, 23.75, 28] #The elements are numbers which includes float as well.
median_data = [30000, 35000, 40000, 35000, 5000000] #The elements are all integers.
mode_data = ['Red', 'Blue', 'Red', 'Red', 'Red'] #The elements are all strings.
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
#Create a figure with multiple subplots (plots inside one figure)
```

plt.subplots usually returns two things:

1. fig: The entire figure (the big canvas).

2. axes: An array of subplot axes (the individual smaller plots).

Breakdown of the arguments:
1, 3 : means 1 row and 3 columns of subplots, so we get 3 plots side-by-side in one row.

figsize=(18, 5): This sets the size of the whole figure to 18 inches wide and 5 inches tall (this makes a wide, short figure).

**MEAN**

```python
mean_data = [20, 22, 25, 23.75, 28]

# ✅ This is correct — creates 1 row, 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

plt.style.use("default")
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3

#Creates a Bar chart. It is colored sky blue.
axes[0].bar(range(len(mean_data)), mean_data, color='skyblue') #Gives x positions and mean_data is the height of each bar.

#Draws a horizontal dashed red line across the plot at the mean value of mean_data.
axes[0].axhline(np.mean(mean_data), color='red', linestyle='--',
                label=f"Mean = {np.mean(mean_data):.2f}") #the label shows the mean value with 2 decimals

#Adds a legend that shows the label for the horizontal mean line
axes[0].legend();
axes[0].set_ylabel("Age") #Labels the y-axis as "Age"
axes[0].set_xticks(range(len(mean_data))) #Sets x-axis ticks at positions [0,1,2,3,4].
axes[0].set_xticklabels([f"P{i+1}" for i in range(len(mean_data))]) #Labels each tick as "P1", "P2", "P3", "P4", "P5" (like Person 1, Person 2, ...).

axes[0].set_title("Mean Imputation Example")#It sets the title of this subplot.

plt.tight_layout()
plt.show()
```

**MEDIAN**

```python
median_data = [30000, 32000, 31000, 30500, 5000000]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

#Creates a Bar chart. It is colored sky blue.
axes[1].bar(range(len(median_data)), median_data, color='lightgreen')

#Adds a horizontal dashed orange line at the median of the income values.
axes[1].axhline(np.median(median_data), color='orange', linestyle='--',
                label=f"Median = {np.median(median_data):,.0f}")
#Label shows the median value (formatted with commas, no decimals)
#Ex: if median = 35000, it displays M=35,000

#Plot a title indicating this is an example of when median is better (due to outliers like 5,000,000).
axes[1].set_title("Median Imputation (Skewed Data)")

axes[1].set_ylabel("Income") #y-axis as "Income"

axes[1].set_xticks(range(len(median_data)))
axes[1].set_xticklabels([f"P{i+1}" for i in range(len(median_data))])
axes[1].legend()

plt.tight_layout()
plt.show()
```

This demonstrates how the median stays stable even when a huge outlier like 5,000,000 is present.

Very effective for teaching why median is better for skewed data.



**MODE**

```python
mode_data = ['Red', 'Blue', 'Red', 'Red', 'Red']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

#Counts how many times 'Red' and 'Blue' appear in the list.
red_count = mode_data.count("Red")
blue_count = mode_data.count("Blue")

#Plots a bar chart with two bars, one for 'Red'and another for 'Blue'.
axes[2].bar(["Red", "Blue"], [red_count, blue_count], color=["crimson", "blue"])

#Setting a title explaining that this plot demonstrates mode imputation for categorical data.
axes[2].set_title("Mode Imputation (Categorical Data)")

#How often each color appears on Y-axis.
axes[2].set_ylabel("Frequency")

#Adds the frequency numbers on top of each bar.
axes[2].text(0, red_count + 0.1, str(red_count), ha='center')
axes[2].text(1, blue_count + 0.1, str(blue_count), ha='center')

plt.tight_layout()
plt.show()
```
